# 주제: <mark> 언론의 프레임 효과 및 정치공학적 분석 </mark>

## 주제 선정 이유 및 목표

<span style="font-size: 12pt"> 최근 대선 기간에 가까워지면서, 많은 사람들이 정치에 이전보다 더 관심을 가지고 있다. 이에 언론이 사람들의 여론을 어떻게 움직이는가에 대한 궁금증이 생겨 관련 내용을 탐구해보고자 한다. 가설은 다음과 같이 세워졌다. </span> 

1. 같은 사건을 보도하더라도 기사의 제목의 어감이나 뉘앙스에 따라 사람들이 접하는 시각이 다르므로, 편향 효과가 발생해 지지율의 변동 정도에 영향을 미칠 것이다.
2. 특정 언론이 어떤 정치인에게 미치는 관심도는 정치인의 지지율에 영향을 미칠 것이다.

<span style="font-size: 12pt"> 가설을 수행하고자, 세운 목표는 다음과 같다; </span> 

1. 언론사별 크롤링을 통해 시기에 따른 기사 제목을 기존의 '감정 단어 사전' 데이터에 접목하여 긍정/부정으로 분류
2. 정치인의 지지율과 기사의 빈도, 어감과의 상관관계 분석
3. 딥러닝을 통해 (1)의 문장의 어감 감지 학습
4. 상관관계 학습을 통해 금달의 기사의 내용에 따라 미리 지지율을 예측

## 과제 수행하면서 느낀 점

1. 가장 크게 느낀 것은 데이터 가공의 중요성이다. 사실 프로젝트를 직접 진행하기 전까지는 우리가 목표를 수행하고자 데이터를 인터넷에서 얻고, 적용하는 과정은 비교적 쉽다고 생각했다. 그러나 가설을 세우고 이에 필요한 데이터는 비슷하게라도 전혀 그대로 찾을 수 없었다. 데이터 마이닝 사이트가 아니라면, 필요한 데이터가 이곳저곳에 흩어져 있거나, 원하는 기준에 따라 정렬되어 있지 않은 경우가 상당히 많았다. 또한, 목표를 명확하게 수행하기 위해서는 데이터를 의도하고자 하는 대로 관리가 필요함을 느꼈다. 조금이라도 정돈되지 못한다면 그 다음 데이터셋을 분석하거나 시각화하는 과정에서 꽤나 오류를 겪음을 느꼈다.


2. 과제를 자기주도적으로 수행하며 딥러닝의 전체적인 컨셉에 대해 알아가게 되어 좋았다. 본 강의를 접하기 전까지는 AI, 딥러닝 등 이러한 용어는 아직까지 추상적이고, 너무 멀고 어려운 개념이라고 느꼈다. 특히 딥러닝은 기계가 계속 연산하면 난제를 모든지 이룰 수 있는 마법 같이 평소에 생각했는데, 딥러닝이란 끊임 없는 기계 학습을 통한 오차 계산이 조금씩 데이터에 대한 전체적인 윤곽을 만드는 기술임을 느끼게 되었다. 딥 러닝 구현 기술은 과제를 하며 많은 난관에 부딪혔는데, 함수의 변수들의 값을 조정해가며 개념에 더 가까워질 수 있어서 흥미로웠다. 방학 때 더 깊이 알고 싶다는 생각이 들었다.


3. 과제를 하면서 가장 아쉬웠던 부분은 데이터 수집에 있었다. 내가 원하는 기준으로 정렬되어 있는 데이터가 없다보니, 웹 크롤러를 활용해서 한정적으로만 데이터를 수집할 수 있었다. 만약 충분히 더 많은 통계가 주어진다면 더 많은 정치인 또는 언론에 대해 분석해 많은 모델을 만들어 가설에 가까워 질 수 있었을 것이라는 아쉬움이 남는다.
